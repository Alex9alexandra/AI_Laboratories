```python
import numpy as np
def spinner(probabilityDistribution):
    r=np.random.random()
    cumulative=0.0
    for i in range(len(probabilityDistribution)):
        cumulative+=probabilityDistribution[i]
        if r<cumulative:
            return i
        
probs=[0.1,0.2,0.4,0.3]
for _ in range(5):
    print(spinner(probs))
```

```python
def spinner2(degrees):
    r=np.random.random()
    cumulative=0
    for i in range(len(degrees)):
        cumulative+=degrees[i]/360
        if r<cumulative:
            return i
dgr=[45,85,70,160]
for _ in range(5):
    print(spinner2(dgr))
```

Mathematically , we take the total area of the rectangle, 10000 cm^2 , and the area of the pipe opening, being pi* 30^2, but, taking into consideration the radius of the balls, the center of the ball needs to be 1 cm away from the wall of the pipe, therefore the area that matters for the pipe is of radius 30-1= 29, so TA=100^2 , SA=pi*29^2, and the probability is the favorable area / total area = SA/TA = 0.264 (26,4%) approximately.

```python
#Monte Carlo simulations:
#we take random events , and the probability is the nmber of favorable events over the total number of events
import numpy as np

#we take parameters
R=0.3
r=0.01
l=1.0
N=1000000 # simulations

#we calculate the effective radius for the ball center 
R_eff=R-r

#we use random.uniform(low,high,size) -> every value in a given range is equally likely to be chosen, numbers are picked between 1 and 0 , and take N random values
x=np.random.uniform(0,1,N)
y=np.random.uniform(0,1,N)
#we have a rectangle of 1mx1m , so the coordinates are between 0 and 1

#take coordinates of pipe, assume it in the middle 
cx=l/2
cy=l/2
#we calculate the distances between the falling ball and the center of the cylinder 
distances=(x-cx)**2+(y-cy)**2
#if that distance is <= the radius ,it means that the ball is inside the pipe
inside=distances<=R_eff**2

prob=np.sum(inside)/N
print(f"Monte Carlo probability: {prob: .4f}")
```

As we can see, the two results are quite close.
